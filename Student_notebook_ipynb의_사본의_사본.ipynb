```python
# Create an empty Canvas
fig =  plt.figure()

# Add the a set of axes to the figure
axes = fig.add_axes([0.1,0.1, 0.8,0.8]) #[left, bottom, width, height] (range 0 to 1)

# plot on the axes
axes.plot(x,y,'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

```python
# Create blank canvas

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # inner axes

#main axes
axes1.plot(x,y,'b')
axes1.set_xlabel('X Label Axis 1')
axes1.set_ylabel('Y Lable Axis 1')
axes1.set_title('Axis 1 Title')


#inner axes
axes2.plot(y,x,'r')
axes2.set_xlabel('X Label Axis 2')
axes2.set_ylabel('Y Lable Axis 2')
axes2.set_title('Axis 2 Title')
```
